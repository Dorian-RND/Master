### Exercice 4

```python
import scipy.cluster.vq as vq
centroid, label = vq.kmeans2(couleur, 3)
print(centroid)
print(label)
```

### Exercice 5

```python
for i in range(len(couleur)):
    couleur[i] = centroid[label[i]]
    
print(couleur)
plt.imshow(np.reshape(couleur, (160,129,3)))
```

### Exercice 6

```python
centres = [couleur[random.randint(0,len(couleur))],couleur[random.randint(0,len(couleur))], couleur[random.randint(0,len(couleur))]]
groupes = []

K = 3

for k in range(K):
        groupes.append([])
        
for i in range(len(couleur)):
    dist = []
    for k in range(K):
        dist.append(np.linalg.norm(centres[k] - couleur[i]))
        groupes[np.argmin(dist)].append(couleur[i])
        
for k in range(K):
    centres[k] = np.mean(groupes[k])
    print("Le centre du cluster ", k, " est : ", centres[k])
        
```
